Написать приложение и функцию, которые собирают основные новости с сайта на выбор dzen.ru, lenta.ru, mail.ru . Для парсинга использовать XPath
Структура данных должна содержать:
* название источника
* наименование новости
* ссылку на новость
* дата публикации

```python
from pprint import pprint
from lxml import html
import requests
```

```python
url = 'https://dzen.ru/news/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

params = {
    'sso_failed': '',
}
```

```python
response = requests.get(url=url, headers=headers, params=params)
response.status_code
```

```python
response.url
```

```python
dom = html.fromstring(response.text)
```

```python
def parser_dzen_n(dom):
    news = dom.xpath("//div[contains(@class,'mg-grid__col_xs')]")
    dzen_news = []
    for new in news:
        news_sourse = new.xpath(".//a[@class='mg-card__source-link']/text()")
        news_text = new.xpath(".//h2/a[@class='mg-card__link']/text()")
        news_link = new.xpath(".//h2/a/@href")
        news_time = new.xpath(".//span[@class='mg-card-source__time']/text()")
        news_dict = {
            'news_sourse': news_sourse,
            'news_text': news_text,
            'news_link': news_link,
            'news_time': news_time
        }
        dzen_news.append(news_dict)
    return dzen_news
```

```python
parser_dzen_n(dom)
```
