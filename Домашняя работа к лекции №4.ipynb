### Задание 2

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
print(t)
print(f)
```

```python
label_font = {
    "fontsize": 9,
    "family": "serif"
}

plt.plot(t, f, color="green")

plt.title("График f(t)", loc="left")
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения f", fontdict=label_font)

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.show()
```

### Задание 3

```python
x = np.linspace(-3, 3, 51)
print(x)
y1 = x ** 2
print(y1)
y2 = 2 * x + 0.5
print(y2)
y3 = -3 * x - 1.5
print(y3)
y4 = np.sin(x)
print(y4)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title("График y1")

ax2.plot(x, y2)
ax2.set_title("График y2")

ax3.plot(x, y3)
ax3.set_title("График y3")

ax4.plot(x, y4)
ax4.set_title("График y4")

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = 'creditcard.csv'
```

```python
df = pd.read_csv(DATASET_PATH, sep=',')
df.head(10)
```

```python
class_l = df["Class"].value_counts()
class_l
```

```python
class_l.plot(kind="bar")
plt.show()
```

```python
class_l.plot(kind="bar")
plt.yscale(value="log")
plt.show()
```

```python
class1 = df.loc[(df["Class"] == 1), "V1"]
```

```python
class0 = df.loc[(df["Class"] == 0), "V1"]
```

```python
plt.hist(class0, bins=20, density=True, alpha=0.5, label="Class 0", color="grey")
plt.hist(class1, bins=20, density=True, alpha=0.5, label="Class 1", color="red")
plt.legend()
plt.xlabel("V1")
plt.show()
```
